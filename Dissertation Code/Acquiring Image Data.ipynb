# Part 1: Acquiring Information from ROI Using Preprocessed MRI Images#

**Import Necessary Packages**

```python
import os
import csv
import pandas as pd
from os import listdir
import numpy as np
import nibabel as nib
from nilearn import datasets

```

**Reading the Nifti Files through all Subfolders**<br>
Here we will be looking into extracting the images from our dataset. Moreover, we will be tallying the records with the challenge dataset which will be used in the  future phase 2 of our model.  

```python
cognition=pd.read_csv('baseline_data.csv')
subject=cognition['Subject']
subject=subject.values.tolist()
count=0
filenames=[]
filename2=[]
subject2=[]
subject3=pd.read_csv('Subjects.csv')


for root, dirs, files in os.walk('G:/ADNI3/'):
    for file in files:
        
        if file.startswith('swlnrADNI'):
            filenames.append(os.path.join(root,file))
            count+=1

    filenames=[w.replace('\\','/') for w in filenames]

subject3=subject3['Subject']
subject3=subject3.values.tolist()

for i in range(len(subject3)):
    if subject3[i] not in subject:
        subject3[i]='Missing'


csvdata=subject3

with open("Directory.csv","w",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerow(csvdata)
csvFile.close()
```

**Acquiring Data from Left Hippocampal Region ROI**<br>
In the below stage, we had acquired the voxel ranges for the Left Hippocampal region of interest. In this stage, we have aquired the intensities along the entire ROI, as well as an estimate to the total number of voxels that accommodate the Left Hippocampal Region. Much of the early formation of Hippocampal region is centered around tissue density(Avg Intensity) and Hippocampal Volume.  

```python
intensity=[]
LHCV=[]

for i in range(len(filenames)):
    intensity.append([])
for i in range(len(filenames)):
    LHCV.append([])
for i in range(len(filenames)):
    example=filenames[i]
    img=nib.load(example)

    data=img.get_fdata()
    
    matrix=[]
    sum=0
    count=0
    rb200=0
    r2400=0
    r4600=0
    r6800=0
    r81000=0
    r12000=0
    for l in range(89,114):
        for j in range(82,119):
            for k in range(40,68):
                if data[l][j][k]<200.0:
                    rb200+=1
                if data[l][j][k]>=200.0 and data[l][j][k]<400.0:
                    r2400+=1
                if data[l][j][k]>=400.0 and data[l][j][k]<600.0:
                    r4600+=1
                if data[l][j][k]>=600.0 and data[l][j][k]<800.0:
                    r6800+=1
                if data[l][j][k]>=800.0 and data[l][j][k]<1000.0:
                    r81000+=1
                if data[l][j][k]>=1000.0 and data[l][j][k]<2000:
                    r12000+=1
                sum=sum+data[l][j][k]
                count+=1
    sum=sum/count
    LHCV[i].append(rb200)
    LHCV[i].append(r2400)
    LHCV[i].append(r4600)
    LHCV[i].append(r6800)
    LHCV[i].append(r81000)
    LHCV[i].append(r12000)
    LHCV[i].sort()
    LHCV[i]=(LHCV[i][-2])
    intensity[i].append(subject3[i])
    intensity[i].append(sum)
    intensity[i].append(LHCV[i])


header=["Subject_ID","Left_ROI_Avg Intensities","Left_Hippocampal_Volume"]
csvdata=header
with open("Left Hippocampus.csv","w",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerow(csvdata)
csvFile.close()

csvdata=intensity
with open("Left Hippocampus.csv","a",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerows(csvdata)
csvFile.close()

LH=pd.read_csv('Left Hippocampus.csv')
print(LH)
```
