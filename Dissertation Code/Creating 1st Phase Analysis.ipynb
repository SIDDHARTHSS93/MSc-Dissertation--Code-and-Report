# Part 2: Gathering Required Information and Analysis of 1st Phase of our Model 

**Importing The necessary Packages**

```python
import csv
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

**Preparing Dataset for 1st Phase Analysis**<br>
The code below, combines the informaion gathered from the Left and Right Hippocampal ROI's prepared in the previous Script. Keeping the Subject ID as common, the features on average intensity and hippocampal volumes are concatenated through columnwise concatenation to form The combined ROI image information tabletable.

```python
df=pd.read_csv('Left Hippocampus.csv')
df1=pd.read_csv('Right Hippocampus.csv')
subject=df['Subject_ID']
df1=df1.drop(['Subject_ID'],axis=1)
df_phase_1=pd.concat([df,df1],axis=1)
df_phase_1=df_phase_1.values.tolist()
header=["Subject_ID","Left_ROI_Avg Intensities","Left_Hippocampal_Volume","Right_ROI_Avg_Intensities", "Right_Hippocampal_Volumes"]
csvdata=header
with open("Combined ROI Image Info.csv","w",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerow(csvdata)
csvFile.close()

csvdata=df_phase_1
with open("Combined ROI Image Info.csv","a",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerows(csvdata)
csvFile.close()

df2=pd.read_csv("Combined ROI Image Info.csv")
print(df2)
```

**Standard Scaler Normalization**<br>
In this stage, we wish to make use of a Dimensionality Reduction, Unsupervised Learning Tool like the Principle Component Analysis function, hence, to do so, we first normalize all the required features.  

```python
df2=pd.read_csv("Combined ROI Image Info.csv")

features=["Left_ROI_Avg Intensities","Left_Hippocampal_Volume","Right_ROI_Avg_Intensities", "Right_Hippocampal_Volumes"]
x=df2.loc[:,features].values
x=StandardScaler().fit_transform(x)
print(x)
```

**PrincipleComponent Analysis**<br>
To the data obtained from the Combined ROI Information. We wish to give a more profound and faster method of analysis. Hence we will make use of the Principle Component Analysis, wherein, the 4Dimensional Image Information(Excluding Subject ID) is reduced to 2 characteristic vectors called eigenvectors, which act as Principle components standardizing the ROI Information from the Left Hippocampus and Right Hippocampus Respectively. Having these 2 vectors, give us the output of our 1st phase in our model.

```python
pca=PCA(n_components=2)
PC=pca.fit_transform(x)

df3=pd.DataFrame(data=PC,columns=['PC1','PC2'])
df3=pd.concat([subject,df3],axis=1)

df3=df3.values.tolist()
header=["Subject_ID","PC1","PC2"]
csvdata=header
with open("1st Phase Output.csv","w",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerow(csvdata)
csvFile.close()

csvdata=df3
with open("1st Phase Output.csv","a",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerows(csvdata)
csvFile.close()

op1=pd.read_csv('1st Phase Output.csv')
print(op1)
```
