**Importing the Necessary Packages**

```python
import pandas as pd
import csv
```

**Acquiring Data from The Challenge Dataset**
In this stage, we are making use of cognitive information that we have acquired from ADNI, based on these Subjects. From the Cognitive information. We will be seeing, factors that might affect the onset of Alzheimers, such as age, gender, education in years, ethnicity, race, Apolipoprotein presence, Mini Mental State Exam Scores. We also acquire the Status variaiable as to whether the patients in this scenario have exhibited a normal aging brain(CN) or an Alzheimer's Disease Affected Brain(AD), or a brain in the Late Stage ofMild Cognitive Impairment(lMCI)

```python
subject=pd.read_csv('Directories.csv')
cognition=pd.read_csv('baseline_data.csv')
subjectc=cognition['Subject']

rid=[]
age=[]
gender=[]
education=[]
ethnicity=[]
race=[]
apoe4=[]
mmsec=[]
data2=[]
ridc=cognition['RID']
agec=cognition['AGE']
genderc=cognition['PTGENDER']
educationc=cognition['PTEDUCAT']
ethnicityc=cognition['PTETHCAT']
racec=cognition['PTRACCAT']
apoe4c=cognition['APOE4']
mmsec=cognition['MMSE']
subject=subject['Subject']
targetc=cognition['DX.bl']
for i in range(len(subject)):
    data2.append([])
for i in range(len(subject)):
    for j in range(len(subjectc)):
        if subject[i]=='Missing':
            data2[i].append(subject[i])
            for j in range(9):
                data2[i].append(None)
            break
        if subject[i]==subjectc[j]:
            data2[i].append(subject[i])
            data2[i].append(ridc[j])
            data2[i].append(agec[j])
            data2[i].append(genderc[j])
            data2[i].append(educationc[j])
            data2[i].append(ethnicityc[j])
            data2[i].append(racec[j])
            data2[i].append(apoe4c[j])
            data2[i].append(mmsec[j])
            data2[i].append(targetc[j])
            break


header=["Subject_ID","RID","AGE","PTGENDER", "PTEDUCATION","PTETHNICITY","PTRACE","APOE4","MMSE","Target"]
csvdata=header
with open("DB_Phase2.csv","w",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerow(csvdata)
csvFile.close()

csvdata=data2
with open("DB_Phase2.csv","a",encoding='utf-8') as csvFile:
    writer=csv.writer(csvFile)
    writer.writerows(csvdata)
csvFile.close()

db2=pd.read_csv('DB_Phase2.csv')
print(db2)
```
